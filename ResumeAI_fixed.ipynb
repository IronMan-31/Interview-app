````python
import re
from docx import Document
from docx.shared import Pt

def markdown_to_docx(markdown_text, filename="resume.docx"):
    doc = Document()
    lines = markdown_text.strip().split('\n')
    inside_code_block = False

    for line in lines:
      line=line.strip()
      print(line)

      if not line and not inside_code_block:
        continue

      if line.startswith("```"):
          inside_code_block = not inside_code_block
          continue

      if inside_code_block:
          doc.add_paragraph(line, style='No Spacing')
          continue

        # HEADINGS
      if line.startswith("#"):
          doc.add_heading(line[1:].strip(), level=0)
      elif line.startswith(" ##") or line.startswith("**"):
          doc.add_heading(line[2:len(line[2:])].strip(), level=1)
      elif line.startswith(" ###") or line.startswith("**"):
          doc.add_heading(line[3:len(line[2:])].strip(), level=2)
      elif line.startswith(" ####"):
          doc.add_heading(line[4:].strip(), level=3)

        # BULLET POINTS
      elif line.startswith("-"):
          bullet_text = line[:].strip()
          bullet_text = re.sub(r'\[([^\]]+)\]\([^)]+\)', r'\1', bullet_text)  # Handle links
          doc.add_paragraph(bullet_text, style='List Bullet')

        # NUMBERED LISTS
      elif re.match(r'^\d+\.\s', line):
          numbered_text = re.sub(r'^\d+\.\s', '', line)
          numbered_text = re.sub(r'\[([^\]]+)\]\([^)]+\)', r'\1', numbered_text)  # Handle links
          doc.add_paragraph(numbered_text, style='List Number')

        # BOLD TEXT (like **Bold Text**)
      elif re.match(r'^\*\*(.*?)\*\*', line):
          bold_text = re.sub(r'\*\*(.*?)\*\*', r'\1', line)
          doc.add_paragraph(bold_text, style='Normal').runs[0].bold = True

        # INLINE FORMATTING (bold, italics, links, etc.)
      else:
            # Handle links, bold, italics, and inline code
          line = re.sub(r'\[([^\]]+)\]\([^)]+\)', r'\1', line)  # Remove link markdown
          line = re.sub(r'\*\*(.*?)\*\*', r'\1', line)  # Remove bold markdown
          line = re.sub(r'_([^_]+)_', r'\1', line)  # Remove italics markdown
          line = re.sub(r'`([^`]+)`', r'\1', line)  # Remove inline code markdown
          doc.add_paragraph(line)

    doc.save(filename)
    return filename
````
